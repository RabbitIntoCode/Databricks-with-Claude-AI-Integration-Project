**Step By Step Process Followed :**

1. Create and set up a Databricks Workspace 

2. Search for the `https://docs.databricks.com/aws/en/dev-tools/cli/install` documnetation adn follow the steps   to install databricks CLI (Used to create any jobs or resources easily with claude)

3. Download and open VS code , create python venv(virtual environment) using  `-m venv.venv` command .Now open a new terminal and the venv should be coming at the start stating that it is in virtual environment . This is to import only necessary packages or bundles from the global python env and libraries.If it's not initiated , type the command in sc to activate it `.venev/Scripts/activate`

![image_1778152419208.png](./image_1778152419208.png "image_1778152419208.png")

4. Now we need to authenticate this workspace in VS code to Dbx.Hence a profile should be created to authenticate to DBx. This is called U2M(user to Machine).

5. To do this go to authenticate tab in DBX CLI help page `https://docs.databricks.com/aws/en/dev-tools/cli/authentication` 

i) The First command to execute will be `databricks auth login --host <workspace-url>` 
for host connection ==> dbx workspace --> SQL warehouses -->select compute-->connection details--> Server host name 
If it doesn't work add `https://` at the start of the host_name

ii) Asks for profile name ==> Give default ot leave it blank and hit enter ==> Redirects to DBX login page , Enter your creds.

We get a message that " profile is successfully saved"(This info will be saved in c drive-->user).this will be used by claude code to send the request/ commands to DBX.

6.Creating the Claude Code: Need to have a Claude Subscription (Costly) , hence we can use API Usage in Claude ,Here we need to add 5 dollars and pay as we go. Search for Claude Code Sign In , we need to buy credits , We can Cretae an API Key (Optional) for our purpose 

7. Download Claude Code ==> search for "Claude Code Docs" --> click on `https://code.claude.com/docs/en/overview` , Use Windows CMD command. We get a Set up Note as shown in SC

![image_1778154450621.png](./image_1778154450621.png "image_1778154450621.png")

Folow the instructions

8. Then Open the terminal again and type `Claude` . the code will be available in terminal

9. To Set up this code Enter `/login` and then select the subscription (1) or API (2).Click on Authorize , the  in terminal Enter `/exit` . At this point the API Key will be automatically created in claude

10. Go to VS code(Restart it) ==> Open a new terminal Enter `claude` (Repeat Step 8 and 9), Instead of Exit now Enter `/model` ==> Select `Sonnet` (medium cost) , `Hiku` (cheap) ==> Give some prompt to test it ==> `/exit` -----This is one time set up.

11. Now search for Databricks AI DEV KIT GITHUB in browser . Opens a github page (this is done by dbx internal field engineering team to boost productivity) `https://github.com/databricks-solutions/ai-dev-kit` Search for Windows default profile and run the command related to it.Select version(`Stable`) --> Select client to use(`Claude Code`) --> default profile --> Give the path to save(`Current Directory`)--> Select Skills to import(`All or DE and AI/ML Engineer`)--> Shows the path where AI Dev kit folder be installed --> and click `Y` to proceed with Installation. Now code folder and MCP json file will appear in our workspace

12. What Exactly happens in background if we provide prompt here ? 

For Ex , if we want to create a Job in dbx , Claude searches for the exact folder in code that we imported , It will have detailed step by step instruction on what exactly need to be done , It opens it and follows the instructions , Similarly there will be specific MCP tool(contains functionality of it) in MCP Json (Can be seen in c drive--> user path that it provide while set up) that it fetches . MCP tool works on the `MCP server` using `mcp_json` and initiates run.

13. Now in terminal start the server again using `claude ` command , To fetch the skills that it has use `/skills` , Give specific prompts like `Create a schema called raw inside databricks catalog called dc_catalog`, It displays it's approach and the objects/tools it's using --> review all those and if everything is fine click `Yes`

14. Now as per our requirement create a prompt like 

For Source(Github) to Bronze layer(DBX) : `Fetch the data from github Urls that are mentioned below and create delta tables on top on these .csv files in dc_catalog.raw schema . Use Python to fetch the data from files and then convert it to pyspark dataframes/concepts while storing in delta tables.Find the URLs below :[give specific urls that have our data (ex : bookings.csv,airports.csv,passengers.csv)]`

It asked for cluster related info : `I suggested it to create a job called Sor_to_bronze.py and then run it to have the tables in target` , We can exit this now using /exit on continue in that terminal only .

For Bronze to Silver : `Join the data in three delta bronze tables(bookings,airports,passengers) in dc_catalog.raw schema based on the join condition into one table (travel_info) in dc_catalog.silver schema. If you do not find schema create it . Join the tables bookings b with passengers p with condition b.passenger_id = p.passenger_id and on airports a with b.booking_id = a.booking_id. Create a new task bronze_to_silver.py`

For Silver to Gold : `Create an aggregated view to count number of bookings per city.Create a new view in dc_catalog.gold schema and Create a new task bronze_to_silver.py`
